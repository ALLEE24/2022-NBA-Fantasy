```python
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
s22 = pd.read_csv('2022.txt')
#Yahoo Fantasy Sports values: Point = 1, Rebound = 1.2, Assist = 1.5, Steal/Block = 3, Turnover= -1
s22['Score/G'] = s22['PTS'] + s22['TRB']*1.2 + s22['AST']*1.5 + s22['BLK']*3 + s22['STL']*3 - s22['TOV']
s22['Score'] = s22['Score/G']*s22['G']

#s22.sort_values(by = ['Score/G'], ascending = False)
```

```python
s21 = pd.read_csv('2021.txt')
s21['Score/G'] = s21['PTS'] + s21['TRB']*1.2 + s21['AST']*1.5 + s21['BLK']*3 + s21['STL']*3 - s21['TOV']
s21['Score'] = s21['Score/G']*s21['G']
s21.sort_values(by = ['Score/G'], ascending = False)
```

```python
#s22.sort_values(by = ['Score'], ascending = False)
```

```python
#Improvement from '21 to '22 in both total and percent
s22['Delta'] = s22['Score'] - s21['Score']
s22['Delta\%'] = s22['Delta']/s21['Score']
s22['Delta/G'] = s22['Score/G'] - s21['Score/G']
s22['Delta\%/G'] = s22['Delta/G']/s21['Score/G']
```
